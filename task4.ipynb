```python
import requests
from bs4 import BeautifulSoup

# URL страницы каталога товаров
url = 'https://aliexpress.ru/category/12/jewelry-accessories?g=y&page=1&searchInfo=A3HzAUR7htIIM+7lIgLXfyyLvu8boisAVpLpeGxtnKNnqF7CLCrT55Rrp1uQpPB1SQcIZFXnOVMcEX6hpzYXdLth8ajd1bQQ0as9wao+TdnWztkVsSU+wBhNanDPtoErANzI6TIRSE6xrGvGa1QvAwb7QoM5JKY7SVf1mGdBWkOlM+3CEiaNoCc3C1bnV3whU5gKxmaNwFEOucrUG0nIj9wLJo0qPXYjff0Knxvy%2FGH7E7%2Fm+g%3D%3D&spm=a2g2w.home.0.0.1d0b5586TH6HIP'

# Заголовки для имитации браузера
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36'
}

# Отправка GET-запроса на страницу каталога
response = requests.get(url, headers=headers)

# Если запрос успешен
if response.status_code == 200:
    # Разбираем HTML-код страницы
    soup = BeautifulSoup(response.text, 'html.parser')

    # Находим все блоки с товарами
    products = soup.find_all('div', class_='red-snippet_RedSnippet__container__1ui42o red-snippet_RedSnippet__vertical__1ui42o')

    if products:
        # Перебираем все товары на странице
        for product in products:
            # Получаем ID товара
            product_id = product['data-product-id']

            # Ищем ссылку на товар
            product_link_element = product.find('a')
            if product_link_element and 'href' in product_link_element.attrs:
                product_link = 'https:' + product_link_element['href']  # Добавляем префикс "https:"
            else:
                product_link = 'Link not found'

            # Ищем название товара прямо на странице каталога
            product_name_element = product.find('div', class_='red-snippet_RedSnippet__title__1ui42o red-typography-utils_style__body_s_regular_16__yl5jcn')
            if product_name_element:
                product_name = product_name_element.get_text(strip=True)
            else:
                product_name = 'Name not found on catalog page'

            # Выводим результаты
            print(f"Product ID: {product_id}")
            print(f"Product Name: {product_name}")
            print(f"Product Link: {product_link}")
            print('-' * 50)  # Разделитель для удобства чтения
    else:
        print("No products found on the page.")
else:
    print("Failed to retrieve the page.")
```
